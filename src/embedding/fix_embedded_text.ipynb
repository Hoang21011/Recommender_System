```python
import numpy as np

embs = np.load(r"D:\food_recommender\embeddings\text_embeddings_cleaned.npy")   # shape: (N, D)
print(embs.shape)
```

```python
# L2 normalize (row-wise)
norms = np.linalg.norm(embs, axis=1, keepdims=True)
embs_norm = embs / norms

# Safety check
print(np.linalg.norm(embs_norm[0]))  # should be ~1.0
```

```python
import faiss

d = embs_norm.shape[1]
index = faiss.IndexFlatL2(d)
index.add(embs_norm.astype("float32"))

print("Vectors in index:", index.ntotal)
```

```python
faiss.write_index(index, "faiss_text_normalized.bin")
```
