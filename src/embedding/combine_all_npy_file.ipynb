```python
import numpy as np

NPY_FILES = [
    r"D:\food_recommender\embeddings\image_embeds_500.npy",
    r"D:\food_recommender\embeddings\image_embeds_500_to_2000.npy",
    r"D:\food_recommender\embeddings\image_embeds_2000_to_4000.npy",
    r"D:\food_recommender\embeddings\image_embeds_4000_to_5000.npy",
    r"D:\food_recommender\embeddings\image_embeds_5000_to_7999.npy",
    r"D:\food_recommender\embeddings\image_embeds_8000_to_9000.npy",
    r"D:\food_recommender\embeddings\image_embeds_9000_to_end.npy",
]
```

```python
import pickle 
```

```python
all_embeds = []

for path in NPY_FILES:
    emb = np.load(path)
    print(f"Loaded {emb.shape} from {path}")
    all_embeds.append(emb)

all_embeds = np.vstack(all_embeds).astype("float32")

print("Final shape:", all_embeds.shape)
```

```python
OUT_PATH = r"D:\food_recommender\embeddings\image_embeds_all.npy"
np.save(OUT_PATH, all_embeds)

print("✅ Saved combined embeddings:", OUT_PATH)
```

```python
with open(r"D:\food_recommender\embeddings\meta_clean_original_paths.pkl", "rb") as f:
    meta = pickle.load(f)

assert all_embeds.shape[0] == len(meta["image_path"])
print("✅ Embeddings perfectly aligned with meta.pkl")
```
